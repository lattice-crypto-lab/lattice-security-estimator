```python
import math
from estimator import *
```

```python
q_dict = {512: {128: 27, 192: 19, 256: 14}, 1024: {128: 54, 192: 37, 256: 29}, 2048: {128: 109, 192: 75, 256: 58}}
```

```python
n = 2**11
expect_security = 100

j_start = 1

for logq in range(106, 138):
    # j = 37
    j = logq - (106 - 40)
    q = 2**logq
    security = 0
    while security < expect_security and j < logq:
        e = 2**j
        param = LWE.Parameters(n=n, q=q, Xs=ND.SparseTernary(n, p=n // 4), Xe=ND.DiscreteGaussian(e))
        try:
            estimator_result = LWE.estimate(
                param,
                deny_list=(
                    "arora-gb",
                    "bkw",
                ),
                jobs=8,
            )
        except:
            print("e=", e)
            print("Error Occur!")
            j = j + 1
        else:
            security = min([math.log(res.get("rop", 0), 2) for res in estimator_result.values()])
            print("n={}".format(n))
            print("logq={}".format(logq))
            print("e=2**{}".format(j))
            print("security={}".format(security))
            print("expect_security={}".format(expect_security))
            if security >= expect_security:
                print("Success!")
                break
            else:
                j = j + 1
```

```python
# n_list = [2**9, 2**10, 2**11]
n_list = [2**9]
# security_list = [128, 192, 256]
security_list = [128]

j_dict = {512: {128: 8, 192: 10, 256: 10}, 1024: {128: 20, 192: 15, 256: 12}, 2048: {128: 40, 192: 30, 256: 25}}
results = {512: {128: [], 192: [], 256: []}, 1024: {128: [], 192: [], 256: []}, 2048: {128: [], 192: [], 256: []}}

for n in n_list:
    for expect_security in security_list:
        logq = q_dict[n][expect_security]
        for i in range(3, -1, -1):
            j = j_dict[n][expect_security]
            q = 2 ** (logq - i)
            security = 0
            while security < expect_security and j < logq - i:
                e = 2**j
                param = LWE.Parameters(n=n, q=q, Xs=ND.SparseTernary(n, p=n // 4), Xe=ND.DiscreteGaussian(e))
                try:
                    estimator_result = LWE.estimate(
                        param,
                        deny_list=(
                            "arora-gb",
                            "bkw",
                        ),
                        jobs=8,
                    )
                except:
                    print("e=", e)
                    print("Error Occur!")
                    j = j + 1
                else:
                    security = min([math.log(res.get("rop", 0), 2) for res in estimator_result.values()])
                    print("n={}".format(n))
                    print("logq={}".format(logq - i))
                    print("e=2**{}".format(j))
                    print("security={}".format(security))
                    print("expect_security={}".format(expect_security))
                    if security >= expect_security:
                        results[n][expect_security].append({"logq": logq - i, "e": e})
                        print("Success!")
                        break
                    else:
                        j = j + 1
```
