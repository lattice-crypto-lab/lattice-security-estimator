```python
from fractions import Fraction
```

```python
def check_noise(n, q, N, logQ, Bg, dg, log_sigma_ks):
    Q = 2**logQ
    sigma_ks = 2**log_sigma_ks

    sigma = 3.19
    left = Fraction(q**2, Q**2) * (
        Fraction(2 * n * dg * N * (Bg**2), 3) * Fraction(sigma**2)
        + Fraction(n * dg * (Bg**2) * (sigma_ks**2), 12)
    ) + Fraction(n // 2 + 1, 12)

    right = Fraction(q**2, (16 * 6) ** 2)
    # print("n={}".format(n))
    # print("q={}".format(q))
    # print("N={}".format(N))
    # print("logQ={}".format(logQ))
    # print("Bg={}".format(Bg))
    # print("dg={}".format(dg))
    # print("log_sigma_ks={}".format(log_sigma_ks))
    # print(left)
    # print(right)
    res = left <= right
    # if res:
    #     print("Valid!")
    # print()
    return res
```

```python
n = 512
N = 2 * n
for q in [n, N]:
    logQ_start = 24
    for logQ in range(logQ_start, 36):
        log_sigma_ks = logQ - (logQ_start - 9)
        max_logBg = 0
        for logBg in range(1, logQ + 1):
            dg = (logQ + logBg - 1) // logBg
            Bg = 2**logBg
            if check_noise(n, q, N, logQ, Bg, dg, log_sigma_ks):
                max_logBg = logBg
            else:
                if max_logBg > 0:
                    # print("n={}".format(n))
                    print("q={}".format(q))
                    # print("N={}".format(N))
                    print("logQ={}".format(logQ))
                    print("logBg={}".format(max_logBg))
                    # print("Bg={}".format(Bg))
                    print("dg={}".format((logQ + max_logBg - 1) // max_logBg))
                    print("log_sigma_ks={}".format(log_sigma_ks))
                    print("Valid!")
                    print()
                break
```
